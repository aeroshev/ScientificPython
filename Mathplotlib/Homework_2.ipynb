### Задание 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
y1, y2, y3, y4
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График y1(x)')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2(x)')

ax3.plot(x, y3)
ax3.set_title('График y3(x)')

ax4.plot(x, y4)
ax4.set_title('График y4(x)')

plt.show()
```

### Задание 4

```python
import pandas as pd
```

```python
from matplotlib import rc

font = {
    'family' : 'fivethirtyeight',
}
rc('font', **font)
```

```python
ccfd = pd.read_csv('creditcard.csv', sep=',')
ccfd.head()
```

```python
fauld_qt = ccfd['Class'].value_counts()
fauld_qt
```

```python
fauld_qt.plot(kind='bar', logy=True)
plt.title('Количество происшействий')
plt.show()
```

```python
not_fauld_v1 = ccfd.loc[ccfd['Class'] == 0, 'V1']
not_fauld_v1
```

```python
fauld_v1 = ccfd.loc[~(ccfd['Class'] == 0), 'V1']
fauld_v1
```

```python
hist_params = {
    'orientation': 'horizontal',
    'ec': 'black',
    'alpha': 0.5,
    'bins': 20,
    'density': True
}
rc('figure', figsize=[12, 6])
```

```python
plt.hist(not_fauld_v1, **hist_params, color='grey', label='Class 0')
plt.hist(fauld_v1, **hist_params, color='red', label='Class 1')
plt.legend(loc='lower right')
plt.xlabel('V1')
plt.show()
```
