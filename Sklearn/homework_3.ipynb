```python
import pandas as pd
import numpy as np
```

### Задание 1

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()

boston.keys()
```

```python
data = boston["data"]
target = boston["target"]
feature_names = boston["feature_names"]
```

```python
X = pd.DataFrame(data, columns=feature_names)
y = pd.DataFrame(target, columns=["price"])
```

```python
X.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

### Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(
    n_estimators=1000,
    max_depth=12,
    random_state=42
)
```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

При анализе результатов тренировки моделей, можно однозначно утверждать, что модель "случайного леса" имеет большую точность предсказания по сравнению с моделью "линейной регресии"

### Задание 4

```python
ccfd = pd.read_csv('../../Материалы/creditcard.csv', sep=',')
ccfd.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
ccfd["Class"].value_counts(normalize=True)
```

```python
pd.options.display.max_columns = 100
```

```python
ccfd.head(10)
```

```python
target = "Class"

X = ccfd.drop(target, axis=1)
y = ccfd[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
parameters = {
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)
}
```

```python
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred_proba_ = clf.predict_proba(X_test)

y_pred_proba = y_pred_proba_[:, 1]

y_pred_proba_[:10]
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred_proba)
```

Очень хороший показатель

```python
from sklearn.datasets import load_wine

wine = load_wine()

wine.keys()
```

```python
data = wine["data"]
target = wine["target"]
target_names = wine["target_names"]
feature_names = wine["feature_names"]
```

```python
print(wine["DESCR"])
```

```python
df = pd.DataFrame(data, columns=feature_names)

df.info()
```

```python
target_df = pd.DataFrame({"target": target}, dtype=np.int64)

df = pd.concat([df, target_df], axis=1)

df.head()
```

```python
df_corr = df.corr()

df_corr
```

```python
import seaborn as sns

sns.set(rc={'figure.figsize':(11.7,8.27)})

hm = sns.heatmap(
    df_corr,
    annot=True
)
```

```python
high_corr = df_corr.loc[
    (df_corr['target'] > 0.5) & (df_corr['target'] < 1),
    :
].index
high_corr
```

```python
df = df.drop("target", axis=1)
```

```python
for feature in high_corr:
    df[f'{feature}_2'] = df[feature]**2

df.head()
```
